Popular Colors

```python
colorlist = []
colortimes = []


for i in range(0,71922269, 1000):
    #print(raw0['color'].iloc[i])
    if raw0['color'].iloc[i] != "NO COLOR":
        if raw0['color'].iloc[i] in colorlist:
            #print("i in colorlist")
            y=colorlist.index(raw0['color'].iloc[i])
            colortimes[y]=colortimes[y] + 1
            #add colortimes
        else:
            #print("i not in colorlist, appended")
            colorlist.append(raw0['color'].iloc[i])
            colortimes.append(1)
```

```python
#sorting both lists
def sort_list(list1, list2):
 
    zipped_pairs = zip(list2, list1)
 
    z = [x for _, x in sorted(zipped_pairs)]
 
    return z
colorlistsorted = sort_list(colorlist, colortimes)
colortimes.sort()
```

```python
#making sure there's a 1:1 ratio between the two lists
colorlistlength=0
colortimeslength=0
for i in colorlist:
    colorlistlength = colorlistlength + 1
for i in colortimes:
    colortimeslength = colortimeslength + 1
print(colorlistlength)
print(colortimeslength)
```

```python
#sorting into array
arr = np.array(colorlistsorted)
arr2 = np.array(colortimes)
#arr2 = [int(numeric_string) for numeric_string in arr2]
arr3 = np.column_stack((arr2,arr))
#sortedArr = arr3[arr3[:,1].argsort()]
#sortedArr = sortedArr[::-1]
#sortedArr = np.argsort(arr3)
#print(arr3)
toptwenty = arr3[-9:-1,:]
```

```python
import matplotlib.pyplot as plt

plt.bar(toptwenty[:,1],toptwenty[:,0], label="Data 1", width=.8, align='center')
```
